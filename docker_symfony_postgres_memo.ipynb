# Docker / Symfony / PostgreSQL – Mémo DogHealth

## 🚀 Docker Compose

```python
# Démarrer tout le projet + Adminer
Docker compose --profile dev up -d --build
```

```python
# Voir les services
Docker compose ps
```

```python
# Logs de tous les services
Docker compose logs -f
```

```python
# Logs API uniquement
Docker compose logs -f dossier
```

```python
# Arrêter
Docker compose stop
```

```python
# Supprimer les conteneurs
Docker compose down
```

```python
# Supprimer conteneurs + volumes
Docker compose down -v
```

```python
# Rebuild complet
Docker compose build --no-cache
```

## 🐘 Symfony / PHP

```python
# Entrer dans le conteneur PHP
Docker compose exec dossier sh
```

```python
# Installer les dépendances
Docker compose exec dossier composer install
```

```python
# Mettre à jour les dépendances
Docker compose exec dossier composer update
```

```python
# Vider le cache
Docker compose exec dossier php bin/console cache:clear
```

```python
# Créer une migration
Docker compose exec dossier php bin/console make:migration
```

```python
# Exécuter les migrations
Docker compose exec dossier php bin/console doctrine:migrations:migrate
```

```python
# Voir les routes
Docker compose exec dossier php bin/console debug:router
```

## 🐘 PostgreSQL

```python
# Entrer dans PostgreSQL
Docker compose exec postgres psql -U $POSTGRES_USER -d $POSTGRES_DB
```

```python
# Shell PostgreSQL
Docker compose exec postgres sh
```

```python
# Vérifier que PostgreSQL répond
Docker compose exec postgres pg_isready
```

### Adminer

URL : http://localhost:8080

- Système : PostgreSQL
- Serveur : postgres
- Utilisateur : POSTGRES_USER
- Mot de passe : POSTGRES_PASSWORD
- Base : POSTGRES_DB
